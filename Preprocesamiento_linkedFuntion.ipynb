 se genera csv con informacion por id de funciones 
 ExecutionTimeDays: el tiempo promedio de execucion durante los 14 dias 
 InvocationsDays: numero de invocaciones promedio durante los 14 dias 
 

los resultados de este procesamiento son usados en KMeans.ipynb

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
def invoc_per_func(ruta = Path.cwd()):
    return  [arch.name for arch in Path(ruta).glob('invocations_per_function_md.anon.d[01][0-9].csv')]

def function_duration(ruta = Path.cwd()):
    return  [arch.name for arch in Path(ruta).glob('function_durations_percentiles.anon.d[01][0-9].csv')]
```

```python
invoc_func = invoc_per_func() 
invoc_per_func= pd.DataFrame()

f_durat = function_duration()
funt_duration= pd.DataFrame()
```

```python
# recorremos archivos function_durations_percentiles para sacar la duracion promedio de cada funcion 
# y luego obtener un promedio general en de los 14 dias 
cont_arch = 1
funt_duration = pd.read_csv("function_durations_percentiles.anon.d01.csv", 
                            usecols=("HashFunction", "Average"))

for arch in f_durat:
    if(arch == "function_durations_percentiles.anon.d01.csv"):
        continue
    else:
        aux = pd.read_csv(arch, usecols =("HashFunction", "Average"))
        funt_duration = funt_duration.merge(aux, on=['HashFunction'], how='outer',suffixes=("_"+str(cont_arch), "_"+str(cont_arch+1)))
        cont_arch+=1
        aux = pd.DataFrame()

cont_arch = 1

funt_duration["ExecutionTimeDays"] = funt_duration.iloc[:, 1:].mean(axis =1)
funt_duration = funt_duration[["HashFunction", "ExecutionTimeDays"]]
funt_duration
```

```python
# recorre invocations_per_function_md.anon para saber el promedio de veces que se invoca cada funcion en los 14 dias 
invoc_per_func = pd.read_csv("invocations_per_function_md.anon.d01.csv")
#dfaux = dfaux. drop(["HashOwner"], axis = 1)
cont_arch = 1
for arch in invoc_func:
    if(arch == "invocations_per_function_md.anon.d01.csv"):
        invoc_per_func["ivocationDay"] = invoc_per_func.iloc[:, 4:].sum(axis=1)
        invoc_per_func = invoc_per_func[["HashFunction", "Trigger", "ivocationDay"]]
        continue                                           
    else:
        aux = pd.read_csv(arch)
        aux["ivocationDay"] = aux.iloc[:, 4:].sum(axis=1)
        aux = aux[["HashFunction", "Trigger", "ivocationDay"]]
        
        invoc_per_func = invoc_per_func.merge(aux, on=['HashFunction','Trigger'], how='outer',suffixes=("_"+str(cont_arch), "_"+str(cont_arch+1)))
        print("van procesadas " + str(cont_arch))
        cont_arch+=1
        aux = pd.DataFrame()
        

invoc_per_func["InvocationsDays"] = invoc_per_func.iloc[:, 2:].mean(axis =1)
invoc_per_func = invoc_per_func[["HashFunction", "Trigger", "InvocationsDays"]]
invoc_per_func

invoc_per_func
```

```python
# poner en columna el tipo de trigger de cada funcion 
copyinvoc = invoc_per_func.copy()
triggers= ["http","event","orchestration","queue","storage","timer","others"] 
for tr in triggers:
    copyinvoc[tr] = np.where(copyinvoc["Trigger"] == tr, 1 , 0)
copyinvoc.InvocationsDays = copyinvoc.InvocationsDays.astype(int)
copyinvoc =  copyinvoc.drop(['Trigger'], axis=1)
copyinvoc
```

```python
Final = pd.merge(funt_duration, copyinvoc, on='HashFunction', how='outer')
Final = Final.dropna().reset_index(drop = True)
Final
```

```python
Final.to_csv('MatrizDataFuntions.csv', index = False)
# archivo con 
```
