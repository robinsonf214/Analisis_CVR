 se genera csv con informacion por id de HashApp 
 generalAverageExecutionTime: el tiempo promedio de execucion durante los 14 dias 
 generalAverageInvocationsDays: numero de invocaciones promedio durante los 14 dias 
 genealAverageMemoy: promedio general utilizado por las aplicaciones en los 12 dias de registro de este parametro
 HashFunction: cantidad de funciones que pertenecen a esa App

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
def invoc_per_func(ruta = Path.cwd()):
    return  [arch.name for arch in Path(ruta).glob('invocations_per_function_md.anon.d[01][0-9].csv')]

def app_memory(ruta = Path.cwd()):
    return  [arch.name for arch in Path(ruta).glob('app_memory_percentiles.anon.d[01][0-9].csv')]

def function_duration(ruta = Path.cwd()):
    return  [arch.name for arch in Path(ruta).glob('function_durations_percentiles.anon.d[01][0-9].csv')]
```

```python
invoc_func = invoc_per_func() 
invoc_per_func= pd.DataFrame()

app_mem = app_memory()
app_memory= pd.DataFrame()

f_durat = function_duration()
funt_duration= pd.DataFrame()
```

```python
# se saca el promedio general por los 12 dias de resgistro para cada HashApp
cont_arch = 1
app_memory = pd.read_csv("app_memory_percentiles.anon.d01.csv", usecols =("HashApp", "AverageAllocatedMb"))

for arch in app_mem:
    if(arch == "app_memory_percentiles.anon.d01.csv"):
        continue
    else:
        aux = pd.read_csv(arch, usecols =("HashApp", "AverageAllocatedMb"))
        app_memory = app_memory.merge(aux, on='HashApp', how='outer',suffixes=("_"+str(cont_arch), "_"+str(cont_arch+1)))
        cont_arch+=1

app_memory["genealAverageMemory"] = app_memory.iloc[:, 1:].mean(axis =1)
app_memory = app_memory[["HashApp","genealAverageMemory"]]
app_memory
```

```python
# recorremos archivos function_durations_percentiles para sacar  promedio general de duracion  en de los 14 dias de cada funcion 
# y mantenemos Hashapp para luego agrupar informacion 

cont_arch = 1
funt_duration = pd.read_csv("function_durations_percentiles.anon.d01.csv", 
                            usecols=("HashApp","HashFunction", "Average"))

for arch in f_durat:
    if(arch == "function_durations_percentiles.anon.d01.csv"):
        continue
    else:
        aux = pd.read_csv(arch, usecols =("HashApp","HashFunction", "Average"))
        funt_duration = funt_duration.merge(aux, on=['HashApp', 'HashFunction'], how='outer',suffixes=("_"+str(cont_arch), "_"+str(cont_arch+1)))
        cont_arch+=1
        aux = pd.DataFrame()

cont_arch = 1
funt_duration["generalAverageExecutionTime"] = funt_duration.iloc[:, 2:].mean(axis =1)
funt_duration = funt_duration[["HashApp", "HashFunction", "generalAverageExecutionTime"]]
funt_duration
```

```python
#sacamos  promedio general de invocaciones de funciones en de los 14 
# y mantenemos Hashapp para luego agrupar informacion 
invoc_per_func = pd.read_csv("invocations_per_function_md.anon.d01.csv")
#dfaux = dfaux. drop(["HashOwner"], axis = 1)
cont_arch = 1
for arch in invoc_func:
    if(arch == "invocations_per_function_md.anon.d01.csv"):
        invoc_per_func["ivocationDay"] = invoc_per_func.iloc[:, 4:].sum(axis=1)
        invoc_per_func = invoc_per_func[["HashApp", "HashFunction", "Trigger", "ivocationDay"]]
        continue
    else:
        aux = pd.read_csv(arch)
        aux["ivocationDay"] = aux.iloc[:, 4:].sum(axis=1)
        aux = aux[["HashApp", "HashFunction", "Trigger", "ivocationDay"]]
        
        invoc_per_func = invoc_per_func.merge(aux, on=['HashApp', 'HashFunction','Trigger'], how='outer',suffixes=("_"+str(cont_arch), "_"+str(cont_arch+1)))
        print("van procesadas " + str(cont_arch))
        cont_arch+=1
        aux = pd.DataFrame()

invoc_per_func["generalAverageInvocationsDays"] = invoc_per_func.iloc[:, 3:].mean(axis =1)
invoc_per_func = invoc_per_func[["HashApp", "HashFunction", "Trigger", "generalAverageInvocationsDays"]]
invoc_per_func
```

acomodamos dataframe's individuales para hacer la agrupacion por HashApp

```python
copyinvoc = invoc_per_func.copy()
countFunc = copyinvoc.drop([ 'generalAverageInvocationsDays', 'Trigger'], axis=1).groupby("HashApp").count()

triggers= ["http","event","orchestration","queue","storage","timer","others"] 
for tr in triggers:
    copyinvoc[tr] = np.where(copyinvoc["Trigger"] == tr, 1 , 0)

countTriger = copyinvoc.drop([ 'generalAverageInvocationsDays', 'HashFunction', 'Trigger'], axis=1).groupby("HashApp").sum()
means =  copyinvoc[['HashApp' ,'generalAverageInvocationsDays']].groupby("HashApp").mean()


invPerFunct = pd.merge(countFunc, means, on='HashApp', how='outer')
invPerFunct = invPerFunct.merge(countTriger, on='HashApp', how='outer') 
invPerFunct.generalAverageInvocationsDays = invPerFunct.generalAverageInvocationsDays.astype(int)


```

```python
funt_duration2 = funt_duration.copy().groupby("HashApp")[["generalAverageExecutionTime"]].mean()
funt_duration2
```

```python
Final = pd.merge(app_memory, funt_duration2, on='HashApp', how='outer')
Final = Final.merge(invPerFunct, on='HashApp', how='outer') 
Final
```

```python
Final = Final.dropna()
Final
```
